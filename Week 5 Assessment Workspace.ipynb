### Step 2

```python
# Create a copy  of the dataframe and clean up the columns by just focusing on the ones you need.
# Consider making a multi-index with both the timeindex and the AthleteID.
columns_to_keep = ['Timestamp', 'Seconds', 'Velocity', 'Acceleration', 'Heart Rate', 'AthleteID']
df = df[columns_to_keep]

# Set the multi-index with Timestamp and AthleteID
df.set_index(['Timestamp', 'AthleteID'], inplace=True)
```

### Step 3

```python
# Make your new variables for the performance metric -- specifically the 20 second farthest distance traveled
# and the 60 second farthest distance traveled

first_20s_data = df[df['Seconds'] <= 20]

first_20s_data['Distance_20s'] = first_20s_data['Velocity'] * first_20s_data['Seconds']

farthest_distance_20s = first_20s_data.groupby(level='AthleteID')['Distance_20s'].max()

first_60s_data = df[df['Seconds'] <= 60]

first_60s_data['Distance_60s'] = first_60s_data['Velocity'] * first_60s_data['Seconds']

farthest_distance_60s = first_60s_data.groupby(level='AthleteID')['Distance_60s'].max()

print("Farthest distance traveled in 20 seconds:")
print(farthest_distance_20s)
print("\nFarthest distance traveled in 60 seconds:")
print(farthest_distance_60s)
```

### Step 4

```python
# Run a groupby with agg to collect your max values for acceleration and your new variables.
agg_dict = {
    'Acceleration': 'max', 
    'Velocity': 'max',     
    'Heart Rate': 'max'     
}

grouped_max_values = df.groupby(level='AthleteID').agg(agg_dict)

grouped_max_values
```

### Step 5

```python
MaxValues_df=df.groupby('AthleteID').agg([max])
print(MaxValues_df)
```

```python
MaxValues_df['zscores_vel'] = stats.zscore(MaxValues_df['Velocity'])
MaxValues_df['zscores_Acc'] = stats.zscore(MaxValues_df['Acceleration'])
MaxValues_df['zscores_HR'] = stats.zscore(MaxValues_df['Heart Rate'])
MaxValues_df
```

### Step 6

```python
constant = 100  # Change this to your desired constant
MaxValues_df['zscores_vel'] += constant
MaxValues_df['zscores_Acc'] += constant
MaxValues_df['zscores_HR'] += constant

# Sum the z-scores to determine performance scores
MaxValues_df['performance_score'] = MaxValues_df['zscores_vel'] + MaxValues_df['zscores_Acc'] + MaxValues_df['zscores_HR']

MaxValues_df

```

### Step 7

```python
# Sort the DataFrame by 'performance_score' column in descending order
MaxValues_df_sorted = MaxValues_df.sort_values(by='performance_score', ascending=False)

# Get the top 2 athletes
top_2_athletes = MaxValues_df_sorted.head(2)

top_2_athletes
```
